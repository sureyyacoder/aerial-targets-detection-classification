```python
import shutil
import os

# Move UAV dataset into the final dataset structure

uav_base_path = "/content/UAV_yolov8"

final_image_dir = "/content/drive/MyDrive/AirAi_Vision/processed/final_dataset/images"
final_label_dir = "/content/drive/MyDrive/AirAi_Vision/processed/final_dataset/labels"

image_paths = [os.path.join(uav_base_path, split, "images") for split in ["train", "valid", "test"]]
label_paths = [os.path.join(uav_base_path, split, "labels") for split in ["train", "valid", "test"]]

# file transfer function
def copy_uav_to_final(image_path, label_path, dest_image_dir, dest_label_dir):
    for img_file in os.listdir(image_path):
        if not img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        img_name = os.path.splitext(img_file)[0]
        label_file = img_name + ".txt"

        img_full = os.path.join(image_path, img_file)
        label_full = os.path.join(label_path, label_file)

        if os.path.exists(label_full):
            shutil.copy(img_full, os.path.join(dest_image_dir, img_file))
            shutil.copy(label_full, os.path.join(dest_label_dir, label_file))

# move all splits one by one
for img_dir, lbl_dir in zip(image_paths, label_paths):
    copy_uav_to_final(img_dir, lbl_dir, final_image_dir, final_label_dir)

print("TaÅŸÄ±ma tamamlandÄ±.")
print("Toplam gÃ¶rsel sayÄ±sÄ± (final_dataset):", len(os.listdir(final_image_dir)))
print("Toplam etiket sayÄ±sÄ± (final_dataset):", len(os.listdir(final_label_dir)))
```
