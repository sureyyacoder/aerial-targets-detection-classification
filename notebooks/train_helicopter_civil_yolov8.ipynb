```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")

model.train(
    data="/content/drive/MyDrive/AirAi_Vision/processed/data_configs/helicopter_civil_data.yaml",
    epochs=30,
    imgsz=640,
    batch=16,
    patience=10,
    augment=True,
    cache=True,
    name="yolov8s_helicopter_civil_30ep",
    project="/content/drive/MyDrive/AirAi_Vision/train_results",
    device=0,
    verbose=True
)
```
