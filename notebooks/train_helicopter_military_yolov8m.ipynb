```python
from ultralytics import YOLO

model = YOLO('yolov8m.pt')  # YOLOv8m = middle level model


model.train(
    data='/content/drive/MyDrive/AirAi_Vision/processed/data_configs/helicopter_military_data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    patience=20,
    name='yolov8m_helicopter_military',
    project='/content/drive/MyDrive/AirAi_Vision/train_results',
    cache=True,
    augment=True,
    verbose=True,
    save_json=True
)
```
