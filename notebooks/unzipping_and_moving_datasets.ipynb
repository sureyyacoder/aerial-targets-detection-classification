```python
import zipfile
import os

#UAV
zip_path = '/content/drive/MyDrive/AirAi_Vision/raw/UAV/UAV_x3_aug_yolov8.zip' # Path of the ZIP file (inside the raw folder on Drive)
extract_path = '/content/UAV_yolov8'  # Path of the extracted folder (directory after unzipping)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("ZIP dosyası açıldı.")
```

```python
#for check
images_path = os.path.join(extract_path, 'train/images')
labels_path = os.path.join(extract_path, 'train/labels')

image_files = sorted([f for f in os.listdir(images_path) if f.endswith(('.jpg', '.png'))])
label_files = sorted([f for f in os.listdir(labels_path) if f.endswith('.txt')])

print(f"Toplam görsel: {len(image_files)}, Etiket: {len(label_files)}")
```

```python
import os
import shutil

# Source base directory
src_base = "/content/UAV_yolov8"

# Destination base directories (for processed images and labels on Google Drive)
dst_base_images = "/content/drive/MyDrive/AirAi_Vision/processed/images"
dst_base_labels = "/content/drive/MyDrive/AirAi_Vision/processed/labels"

splits = ['train', 'valid', 'test']

for split in splits:
    # Define destination folder paths (for images and labels)
    image_dst = os.path.join(dst_base_images, split, "UAV")
    label_dst = os.path.join(dst_base_labels, split, "UAV")
    os.makedirs(image_dst, exist_ok=True)
    os.makedirs(label_dst, exist_ok=True)

    # Define source folder paths (for images and labels)
    src_images = os.path.join(src_base, split, "images")
    src_labels = os.path.join(src_base, split, "labels")

    # Copying operation
    for f in os.listdir(src_images):
        shutil.copy(os.path.join(src_images, f), os.path.join(image_dst, f))

    for f in os.listdir(src_labels):
        shutil.copy(os.path.join(src_labels, f), os.path.join(label_dst, f))

print("Yeni UAV veri seti başarıyla taşındı ve klasör yapısı oluşturuldu.")
```

```python
yaml_content = """
train: /content/drive/MyDrive/AirAi_Vision/processed/images/train/UAV
val: /content/drive/MyDrive/AirAi_Vision/processed/images/valid/UAV
test: /content/drive/MyDrive/AirAi_Vision/processed/images/test/UAV

nc: 1
names: ['UAV']
"""

with open("/content/drive/MyDrive/AirAi_Vision/processed/data_configs/uav_data.yaml", "w") as f:
    f.write(yaml_content.strip())

print("data.yaml başarıyla yazıldı.")
```
