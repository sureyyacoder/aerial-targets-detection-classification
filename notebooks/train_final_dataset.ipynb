```python
from ultralytics import YOLO

#  Modeli yükle
model = YOLO('yolov8m.pt')  # YOLOv8 Medium model

#  Eğitim
results = model.train(
    data='/content/drive/MyDrive/AirAi_Vision/processed/final_dataset_split/data.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    patience=20,
    name='yolov8m_final_dataset_100ep_v1',
    project='/content/drive/MyDrive/AirAi_Vision/train_results',
    cache=True
)

print("Eğitim tamamlandı. Sonuç klasörü:", results.save_dir)
```
